Twitter Trends
==============

### Connect to Twitter:

```python
credentials = yaml.load(open(os.path.expanduser('~/api_cred.yml')))

auth = twitter.oauth.OAuth(credentials['ACCESS_TOKEN'], 
                           credentials['ACCESS_TOKEN_SECRET'],
                           credentials['API_KEY'],
                           credentials['API_SECRET'])

twitter_api = twitter.Twitter(auth=auth)
```

```python
# The Yahoo! Where On Earth ID for the entire world is 1.
# See https://dev.twitter.com/docs/api/1.1/get/trends/place and
# http://developer.yahoo.com/geo/geoplanet/

WORLD_WOE_ID = 1
US_WOE_ID = 23424977

# Prefix ID with the underscore for query string parameterization.
# Without the underscore, the twitter package appends the ID value
# to the URL itself as a special case keyword argument.

world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)

pprint(world_trends)
```

```python
pprint(us_trends)
```

### Do any of these trends intersect?

```python
world_trends_set = set([trend['name'] for trends in world_trends
                        for trend in trends['trends']])

us_trends_set = set([trend['name'] for trends in us_trends
                     for trend in trends['trends']]) 

world_trends_set.intersection(us_trends_set)
```

```python
# Loop through the list object to create Pandas dataframe that contains target data from each record

pop_users_canada_df = pd.DataFrame({'followers_count': tweet['user'].get('followers_count'), 
                                    'user_name': tweet['user'].get('name')} 
                                   for tweet in pop_users_canada)

pop_users_canada_df.head()
```

```python
pop_users_usa_df = pd.DataFrame({'followers_count': tweet['user'].get('followers_count'), 
                                    'user_name': tweet['user'].get('name')} 
                                   for tweet in pop_users_usa)
```
